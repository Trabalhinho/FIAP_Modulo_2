```python
def categorizar_tomate_media( media ):
  if media < 40:
    return 'Tomate pequeno'
  elif media >= 40 and media <= 70:
    return 'Tomate mdio'
  else:
    return 'Tomate grande'
```

```python
dados_tomato['categoria_tomate'] = dados_tomato['Average'].apply( categorizar_tomate_media )
dados_tomato.head()
```

```python
dados_tomato.groupby( ['categoria_tomate'] ).describe().T
```

```python
filtro = dados_tomato['Average'] < 40
dados_tomato.loc[ filtro ]
```

```python
from sklearn.model_selection import train_test_split
```

```python
dados_diabetes = pd.read_csv( 'https://raw.githubusercontent.com/Trabalhinho/FIAP_Modulo_2/main/Base_de_dados/diabetes.csv', sep = ',' )
dados_diabetes.head()
```

```python
x = dados_diabetes.drop( ['Class variable'], axis = 1 )
x.head()
```

```python
y = dados_diabetes['Class variable']
```

###Aula 9

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size = 0.3 )
```

```python
len( x_train )
```

```python
len( x_test )
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier( n_neighbors = 3 )
```

```python
knn.fit( x_train, y_train )
```

```python
accuracy = knn.score( x_test, y_test )
```
