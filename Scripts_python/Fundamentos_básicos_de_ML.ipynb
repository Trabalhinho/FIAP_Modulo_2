```python
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
dados_diabetes = pd.read_csv( 'https://raw.githubusercontent.com/Trabalhinho/FIAP_Modulo_2/main/Base_de_dados/diabetes.csv', sep = ',' )
dados_diabetes.head()
```

```python
x = dados_diabetes.drop( ['Class variable'], axis = 1 )
x.head()
```

```python
y = dados_diabetes['Class variable']
```

###Aula 2

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size = 0.3 )
```

```python
len( x_train )
```

```python
len( x_test )
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier( n_neighbors = 3 )
```

```python
knn.fit( x_train, y_train )
```

```python
accuracy = knn.score( x_test, y_test )
```

###Aula 8

```python
def classifica_musica_ao_vivo( df ):
  if df['liveness'] >= 0.8:
    return True
  else:
    return False
```

```python
dados_rolling_stones['ao_vivo'] = dados_rolling_stones.apply( classifica_musica_ao_vivo, axis = 1 )
dados_rolling_stones.head()
```

```python
dados_rolling_stones.groupby( 'ao_vivo' )['ao_vivo'].count()
```

```python
dados_rolling_stones_gravado_em_estudio = dados_rolling_stones[ dados_rolling_stones['ao_vivo'] == False ]
dados_rolling_stones_show_ao_vivo = dados_rolling_stones[ dados_rolling_stones['ao_vivo'] == True ]
```

```python
print( 'Média das músicas ao vivo: ', dados_rolling_stones_show_ao_vivo['duration_minutes'].mean() )
```

```python
print( 'Média das músicas gravadas em estúdio: ', dados_rolling_stones_gravado_em_estudio['duration_minutes'].mean() )
```

###Aula 9

```python
dados_studio = dados_rolling_stones_gravado_em_estudio.groupby('album')['loudness'].sum()
dados_ao_vivo = dados_rolling_stones_show_ao_vivo.groupby('album')['loudness'].sum()

fig, axes = plt.subplots( 1, 2, figsize = (15, 4) )

sns.histplot( data = dados_studio, bins = 20, ax = axes[0] )
axes[0].set_title('Som do barulho no estúdio')
axes[0].set_xlabel('Barulho')
axes[0].set_ylabel('Frequência')

sns.histplot( data = dados_ao_vivo, bins = 20, ax = axes[1] )
axes[1].set_title('Som do barulho ao vivo')
axes[1].set_xlabel('Barulho')
axes[1].set_ylabel('Frequência')

fig.tight_layout()
plt.show()
```

###Aula 10

```python
plt.figure( figsize = (10, 5) )

sns.kdeplot( data = dados_studio, label = 'Álbuns de estúdio', fill = True )
sns.kdeplot( data = dados_ao_vivo, label = 'Álbuns ao vivo', fill = True )

plt.title( 'Distribuição do barulhos dos álbuns' )
plt.xlabel('Barulho')
plt.ylabel('Densidade')
plt.legend()

plt.show()
```

```python
from scipy.stats import shapiro
```

```python
stat, p = shapiro( dados_studio )
print( 'Soma do barulho dos álbuns de estúdio: ' )
print( 'Estatística de teste: {:.4f}, valor p: {}'.format(stat, p) )

if p > 0.05:
  print( 'Não há evidência para rejeitar a hipótese de normalidade' )
else:
  print( 'A hipótese de normalidade é rejeitada' )
```

```python
stat, p = shapiro( dados_ao_vivo )
print( 'Soma do barulho dos álbuns ao vivo: ' )
print( 'Estatística de teste: {:.4f}, valor p: {}'.format(stat, p) )

if p > 0.05:
  print( 'Não há evidência para rejeitar a hipótese de normalidade' )
else:
  print( 'A hipótese de normalidade é rejeitada' )
```

###Aula 11

```python
from scipy.stats import mannwhitneyu
```

```python
stat, p = mannwhitneyu( dados_studio.sample( len( dados_studio ) ), dados_ao_vivo.sample( len( dados_ao_vivo ) ), alternative = 'less' )

print( 'Estatística de teste U: ', stat )
print( 'Valor p: ', p )

alpha = 0.05
if p < alpha:
  print( 'Diferença estatisticamente significante' )
else:
  print( 'Não há diferença estatisticamente significante' )
```

###Aula 12

```python
media_por_album = dados_rolling_stones.groupby('album')['valence'].mean().reset_index()

media_por_album = media_por_album.rename( columns = {'valence': 'media_valence'} )

media_por_album['sentimento'] = ['positivo' if v > 0.6 else 'negativo' for v in media_por_album['media_valence'] ]

media_por_album.groupby('sentimento')['sentimento'].count()
```

```python
dados_resultado_final = pd.merge( dados_rolling_stones, media_por_album, on = 'album' )
```

###Aula 13

```python
matriz_correlacao = dados_resultado_final.corr()

correlacao_sentimento = matriz_correlacao['media_valence']

display(correlacao_sentimento)
```

```python
sns.heatmap( correlacao_sentimento.to_frame(), annot = True, cmap = 'coolwarm' )
plt.show()
```

```python
sns.scatterplot( x = 'media_valence', y = 'danceability', hue = 'sentimento', data = dados_resultado_final, palette = 'coolwarm' )
plt.xlabel('Média valence')
plt.ylabel('Danceability')
plt.title('Relação entre a valência, média e capacidade de dança das músicas')

plt.show()
```

```python
sns.scatterplot( x = 'media_valence', y = 'liveness', hue = 'sentimento', data = dados_resultado_final, palette = 'coolwarm' )
plt.xlabel('Média valence')
plt.ylabel('Liveness')
plt.title('Relação entre a valência, média e se a música é ao vivo músicas')

plt.show()
```
